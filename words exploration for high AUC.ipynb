For words that occur atleast n (n=5?) time within the project: 

Split into two categories 
- a) words that also occur in atleast m (m=1 ?) other projects
- b) words that occur only in this project.

```python
import pandas as pd
import pickle
import numpy as np
import nltk
import re
import itertools
import bisect
from datetime import datetime
from nltk.tokenize.casual import TweetTokenizer
from nltk.collocations import TrigramCollocationFinder
from operator import itemgetter
import argparse, os, csv, sys, hashlib, pickle, random
from collections import Counter
```

```python
with open("comments_dict_raw.txt",'rb') as f:
    commentdata = pickle.load(f)
```

```python
projectname = []
groups = dict()
for key, group in itertools.groupby(commentdata, key=lambda x:x['project']):
    projectname.append(key)
    groups[key] = list(group)
```

```python
groups.keys()
```

```python
grpedcmts = dict()

for key in groups.keys():
    tk = TweetTokenizer(preserve_case=True, reduce_len=False, strip_handles=False)
    grpedcmts[key] = Counter(tk.tokenize(' '.join([re['tk_body'] for re in groups[key]])))

```

```python
mullist = []
for key in groups.keys():
    mullist = mullist + [k for k,v in grpedcmts[key].items() if v>=50]
```

```python
commonlist = [k for k,v in Counter(mullist).items() if v>=30]
p = re.compile('\d+(\.\d+)?')
commonlist = [k for k in commonlist if p.match(k)==None]
```

### 178 common words!

```python
commonlistdesc = pd.DataFrame(columns = list(groups.keys()),index = commonlist)
commonlistdesc.columns
commonlistdesc.loc['I','zxing_zxing']

for key in groups.keys():
    for word in commonlist:
        commonlistdesc.loc[word,key] = grpedcmts[key][word]/sum(grpedcmts[key].values())
```
